```python
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("0.csv", header=None)
df
```

```python
img = cv2.imread("0.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

for i in df.iterrows():
  img = cv2.rectangle(img, (i[1][1], i[1][2]), (i[1][3], i[1][4]), (255,0,0), 2)


plt.imshow(img)
```
